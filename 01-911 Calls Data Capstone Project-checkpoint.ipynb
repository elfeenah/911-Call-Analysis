## Creating new features

** In the titles column there are "Reasons/Departments" specified before the title code. These are EMS, Fire, and Traffic. Use .apply() with a custom lambda expression to create a new column called "Reason" that contains this string value.** 

**For example, if the title column value is EMS: BACK PAINS/INJURY , the Reason column value would be EMS. **

```python
df['Reason'] = df['title'].apply(lambda x: x.split(':')[0])
```

** Now let us begin to focus on time information. What is the data type of the objects in the timeStamp column? **

```python
timeStamp_data_type = df['timeStamp'].dtype
print(timeStamp_data_type)
```

** You should have seen that these timestamps are still strings. Use [pd.to_datetime](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.to_datetime.html) to convert the column from strings to DateTime objects. **

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

** You can now grab specific attributes from a Datetime object by calling them. For example:**

    time = df['timeStamp'].iloc[0]
    time.hour

**You can use Jupyter's tab method to explore the various attributes you can call. Now that the timestamp column are actually DateTime objects, use .apply() to create 3 new columns called Hour, Month, and Day of Week. You will create these columns based off of the timeStamp column, reference the solutions if you get stuck on this step.**

```python
df['Hour'] = df['timeStamp'].apply(lambda time: time.hour)
df['Month'] = df['timeStamp'].apply(lambda time: time.month)
```

```python
df['Day of Week'] = df['timeStamp'].apply(lambda time: time.dayofweek)
```

** Notice how the Day of Week is an integer 0-6. Use the .map() with this dictionary to map the actual string names to the day of the week: **

    dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}

```python
dmap = {0: 'Mon', 1: 'Tue', 2: 'Wed', 3: 'Thu', 4: 'Fri', 5: 'Sat', 6: 'Sun'}
```

```python
df['Day of Week'] = df['Day of Week'].map(dmap)
```

**Create a new column called 'Date' that contains the date from the timeStamp column. You'll need to use apply along with the .date() method. **

The figure shows the counts of 911 calls over time, grouped by date. Each point on the plot represents a specific date, and the height of the point represents the count of 911 calls on that date. The plot allows us to visualise the overall trend and patterns in the number of 911 calls over time. We can observe any spikes or dips in call volume and identify any recurring patterns or anomalies. This information can be useful for understanding the demand for emergency services and identifying potential areas for improvement or resource allocation.

** Now recreate this plot but create 3 separate plots with each plot representing a Reason for the 911 call**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create separate dataframes for each reason
ems_df = df[df['Reason'] == 'EMS']
fire_df = df[df['Reason'] == 'Fire']
traffic_df = df[df['Reason'] == 'Traffic']

# Group by date and count the number of calls for each reason
ems_counts = ems_df.groupby('Date').count()
fire_counts = fire_df.groupby('Date').count()
traffic_counts = traffic_df.groupby('Date').count()

# Reset the index to make Date a column
ems_counts = ems_counts.reset_index()
fire_counts = fire_counts.reset_index()
traffic_counts = traffic_counts.reset_index()

# Create subplots
fig, axes = plt.subplots(nrows=3, figsize=(10, 8))

# Plot for EMS calls
sns.lineplot(x='Date', y='e', data=ems_counts, ax=axes[0])
axes[0].set_title('EMS Calls')

# Plot for Fire calls
sns.lineplot(x='Date', y='e', data=fire_counts, ax=axes[1])
axes[1].set_title('Fire Calls')

# Plot for Traffic calls
sns.lineplot(x='Date', y='e', data=traffic_counts, ax=axes[2])
axes[2].set_title('Traffic Calls')

# Adjust spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

The figure displays three separate line plots representing the count of 911 calls over time for different reasons: EMS, Fire, and Traffic. Each plot corresponds to a specific reason and shows the trend in call volumes. The plots enable easy comparison and analysis of the temporal patterns in emergency incidents for each category.

```python
# Plot for Fire calls
sns.lineplot(x='Date', y='e', data=fire_counts, ax=axes[1])
axes[1].set_title('Fire Calls')
```

```python
# Plot for Traffic calls
sns.lineplot(x='Date', y='e', data=traffic_counts, ax=axes[2])
axes[2].set_title('Traffic Calls')
```

** Now let's move on to creating  heatmaps with seaborn and our data. We'll first need to restructure the dataframe so that the columns become the Hours and the Index becomes the Day of the Week. There are lots of ways to do this, but I would recommend trying to combine groupby with an [unstack](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.unstack.html) method. Reference the solutions if you get stuck on this!**

```python
heatmap_data = df.groupby(['Day of Week', 'Hour']).count()['Reason'].unstack(level='Hour')
```

** Now create a HeatMap using this new DataFrame. **

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap='viridis')
plt.title('911 Calls Heatmap by Day of Week and Hour')
plt.xlabel('Hour')
plt.ylabel('Day of Week')
plt.show()
```

The heatmap visually represents the frequency of 911 calls based on the day of the week and hour. It shows that weekdays, particularly during daytime hours, have the highest call volumes. Weekends have lower call volumes, distributed more evenly throughout the day. The heatmap helps identify peak periods and patterns in call frequency for resource allocation and planning.
